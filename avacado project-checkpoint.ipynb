```python
features = df.drop(columns = ['AveragePrice','region'],axis=1)
labels = df['AveragePrice']
X = features
Y = labels
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 2)
```

```python
lr=LinearRegression()
lr.fit(X_train,Y_train)
pred=lr.predict(X_test)
```

```python
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(Y_test, pred))
print('MSE:', metrics.mean_squared_error(Y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, pred)))
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr=DecisionTreeRegressor()
dtr.fit(X_train,Y_train)
pred=dtr.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(Y_test, pred))
print('MSE:', metrics.mean_squared_error(Y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, pred)))
```

```python
rdr = RandomForestRegressor()
rdr.fit(X_train,Y_train)
pred=rdr.predict(X_test)
```

```python
print('MAE:', metrics.mean_absolute_error(Y_test, pred))
print('MSE:', metrics.mean_squared_error(Y_test, pred))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, pred)))
```

Well as we can see the RMSE is lower than the two previous models, so the RandomForest Regressor is the best model in this case.


NOW WE WILL TAKE THE TARGET AS REGION 

```python
features = df.drop(columns = ['region'],axis=1)
labels = df['region']
X = features
Y = labels
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size= 0.2, random_state = 2)
```

```python
dc = DecisionTreeClassifier()
```

```python
dc.fit(X_train,Y_train)
```

```python
y_pred= dc.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred))

```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
```

```python
knn.fit(X_train,Y_train)
```

```python
y_pred_1= knn.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(Y_test, y_pred_1))
```

```python
from sklearn.model_selection import StratifiedKFold,cross_val_score
skf = StratifiedKFold(n_splits=10,shuffle=True,random_state=10)

print(f'KNeighbors Classifier Accuracy is {round((cross_val_score(knn,X,Y,cv=skf,scoring="accuracy").mean())*100,2)}')
print(f'Decision Tree Classifier Accuracy is {round((cross_val_score(dc,X,Y,cv=skf,scoring="accuracy").mean())*100,2)}')
```

AS WE CAN CHECK FROM THE ABOVE THAT DECISION TREE IS THE BEST ALGORITHM FOR  REGION AS TARGET VARIABLE.
